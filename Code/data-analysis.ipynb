```python
# Import Dependencies
import pandas as pd 
import numpy as np
```

```python
# Import CSV
# CSV Source: https://ourworldindata.org/grapher/oil-consumption-by-country?tab=table
df_oil = pd.read_csv('../Data/oil-consumption-by-country.csv')
df_oil.tail(20)
```

```python
# Filter the World data only
df_oil_world = df_oil.loc[df_oil['Entity'] == 'World'][['Year','Oil consumption - TWh']]
df_oil_world = df_oil_world.set_index('Year')
df_oil_world = df_oil_world.iloc[-23:] # Grab from 2000 and onward
df_oil_world.head(23)

# Save data
df_oil_world.to_csv('../Data/Oil Consumption.csv', index=False)
```

```python
# Create nice looking chart
fig = df_oil_world.plot(ylabel="TeraWatts Hour", xticks=np.arange(2000,2021,1), grid=True, figsize=(40,10))

# Save Figure
fig.figure.savefig('../Chart/Oil Consumption Chart.png')
```
